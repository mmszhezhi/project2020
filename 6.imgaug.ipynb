```python
import imageio
import imgaug as ia
from imgaug import augmenters as iaa
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
image = imageio.imread("https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png")

print("Original:")
ia.imshow(image)
```

```python
print(type(image))
```

```python
ia.seed(4)
rotate = iaa.Affine(rotate=(-180, 180))
crop = iaa.Crop(percent=(0, 0.3))
mul = iaa.Multiply((0.5,3.5), per_channel=0.01)
f,ax = plt.subplots(4,6,figsize=(16,9))
for i in range(4):
    for j in range(6):   
        image_aug = mul(image=image)
        ax[i][j].set_title(image_aug.shape)
        ax[i][j].set_xticks([])
        ax[i][j].set_yticks([])
        
        ax[i][j].imshow(image_aug)
```

```python
import numpy as np

images = [image, image, image, image]
rotate = iaa.Affine(scale=0.3)
images_aug = rotate(images=images)

print("Augmented batch:")
ia.imshow(np.hstack(images_aug))
```
